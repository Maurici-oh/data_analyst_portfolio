```python
#Import relevant libraries
import pandas as pd
from datetime import datetime
```

```python
# Load raw data from the CSV file
raw_data = pd.read_csv('eu_uk_inflation_2015_2025.csv')
df = raw_data.copy()
```

```python
# Display the first rows of the data
df.head()
```

```python
# Select the relevant columns
df_filtered = df[['coicop','geo','TIME_PERIOD','OBS_VALUE']]
df_filtered.reset_index(drop=True)
df_filtered
```

```python
# Convert the "date" column to actual datetime data type
df_filtered['date'] = pd.to_datetime(df['TIME_PERIOD'], format='%Y-%m')
df_filtered.head()
```

```python
# Drop the original date column
df_eu = df_filtered.drop(['TIME_PERIOD'], axis=1)
df_eu.head()
```

```python
# Double check the column's data types
df_eu.dtypes
```

```python
# Print the amount of rows and columns
df_eu.shape
```

```python
# Export the data frame as Excel file
df_eu.to_excel('Eu_Inflation_Trends_Analysis.xlsx', index=False)
```
